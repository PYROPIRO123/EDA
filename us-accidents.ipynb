2. START TIME

```python
#records for when all accidents happened
df.Start_Time
```

```python
#we are converting the dtype to datetime from object
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0].hour
```

```python
#records for at what hour accidents happened
df.Start_Time.dt.hour
```

```python
#plotting a distribution curve for the same
sns.distplot(df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
```

- A high percentage of accidents occur between 2 pm to 5 pm. The data shows that the rise happends after noon
- Although people commute to work in the morning, the accidents that happen during that time is nominal when compared to 2-5 pm

```python
#this plot shows the data curve for all the days of the week
sns.distplot(df.Start_Time.dt.dayofweek, bins=24, kde=False, norm_hist=True)
```

- On weekdays the more accidents happen, especially on FRIDAY i.e. when people have the end of work day and pubs open

```python
saturdays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 5]
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
```

```python
sns.distplot(sundays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

-On Sundays the peak occurs at around 2 pm till 5 pm, almost like weekdays

```python
sns.distplot(saturdays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

Trend - seems to be more accidents happening in december or winter maybe coz there are more people outside on the street

```python
#for month the sundays analysis
sns.distplot(sundays_start_time.dt.month, bins=12, kde=False, norm_hist=True)
```

```python
df_2019 = df[df.Start_Time.dt.year == 2019]
sns.distplot(df_2019.Start_Time.dt.month, bins= 12, kde=False, norm_hist=True)
```

5. Weather Condition

```python
#well the weather condition is textual in fair, cloudy, etc.
df.Weather_Condition
df_weather_condition = df.Weather_Condition.value_counts()
df_weather_condition
```

SUMMARY AND INSIGHTS:
- No Data from New York
- Less than 5% of cities have more than 1000 yearly accidents
- The number of accidents per city decreases exponentially
- Over 1100 cities have reported just one accident (need more investigation)
- Mostly in the range of 55-75 F the accidents happen, also in Fair weather, which means although cold and low visibility should cause more accidents and it is, but the peak is in normal fair weather
